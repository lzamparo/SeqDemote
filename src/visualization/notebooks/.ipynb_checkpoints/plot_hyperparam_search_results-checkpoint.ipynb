```python
import os
import re
import plotnine as gg
import pandas as pd

results_dir = '/Users/zamparol/projects/SeqDemote/results/spearmint_runs/pytorch_models'

# Best parameters {'sparse_lambda': 1e-08, 'weight_lambda': 0.09957235087741835, 'orth_lambda': 1.0, 'bias_lambda': 0.00010165984461459264}
def best_line_to_dict(l, pattern):
    ''' strip the best params, return as a dict'''
    param_name, param_val = [],[]
    l = l.strip()
    matches = re.findall(pattern, l)
    for m in matches:
        k, v = m
        param_name.append(k)
        param_val.append(float(v))
    
    return param_name, param_val
    
def strip_best_params(filename, pattern):
    ''' get the best params, turn them into a dict '''
    with open(filename,'r') as f:
        for line in f:
            if not line.startswith("Best"):
                continue
            experiment_params, experiment_vals = best_line_to_dict(line, pattern)
    return experiment_params, experiment_vals

kv_pattern = re.compile("'([\w]+)': ([\d|e|\-|.]+)")    

# get the data
results_files = [f for f in os.listdir(results_dir) if f.endswith(".stdout")]
```

```python
param_names, param_vals = [], []
for f in results_files:
    names, values = strip_best_params(os.path.join(results_dir,f), kv_pattern)
    param_names.extend(names)
    param_vals.extend(values)
```

```python
df = pd.DataFrame.from_dict({'hyperparam': param_names, 'value': param_vals})
df
```

```python
%matplotlib inline
p = gg.ggplot(df, gg.aes('hyperparam','value'))
p = p + gg.geom_jitter(width=0.05)
p
```
